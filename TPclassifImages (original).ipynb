```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
sift = cv2.SIFT_create()
now = time.time()
kp, des = sift.detectAndCompute(gray,None)
duration = time.time()-now
print("durée = %.2f s" %duration)
cv2.drawKeypoints(gray, kp, img, flags = cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(gray, cmap='gray')
```

## Chargement de vos classes d'images

```python
baseTrain = basedir+'train/'
classes = ['pecora','ragno']

lesSifts = np.empty(shape=(0, 128), dtype=int)
dimImgs = []

nbImages = 0
for cl in classes:
    nbImages += len(glob.glob(baseTrain+cl+'/*'))
#xTrain = np.empty(shape=(nbImages,size2))
print("total number of images: ", nbImages)
i = 0
nb = [] # pour stocker le nombre d'images de chaque classe

for cl in classes:
    now = time.time()
    listImages = glob.glob(baseTrain+cl+'/*')
    print("class: ", cl, " : ", len(listImages), " images.") 
    nb.append(len(listImages))
    for pathImg in tqdm(listImages, position=0, leave=True): # POUR DEBUGGUER, VOUS POUVEZ REDUIRE LE NB D'IMAGES!!
        img = io.imread(pathImg)
        
    duration = time.time()-now
    print('classe {}: {:.2f} s'.format(cl, duration))
```
